This file iterates over the available data and combines it into a clean combined training file. More precise preprocessing is handled in the training script; this script simply extracts the body of the target texts and groups them into a single file.

```python
parse_file = open("parse_list.txt", 'r')

parsing_dict = {}
for line in parse_file:
  if line[0] == '#':
    pass  # skip comments
  else:
    subject, start, end = line.strip().split(' ')
    parsing_dict[subject] = (int(start), int(end))

parsing_dict
```

```python
corpus = ""

for (subject, (start, end)) in parsing_dict.items():
  filename = subject + ".txt"
  f = open(filename, 'r', encoding='ISO-8859-1')
  lines = f.readlines()
  # the line numbers in parsing_dict are indexed from 1
  text_body = ''.join(lines[start-1:end])
  corpus += text_body + '\n\n'
  f.close()
```

```python
corpus = corpus.replace("----", "--")
corpus = corpus.replace("---", "--")

# In the final language model, individual hyphens are treated as tokens. Hyphens
# also tend to follow one another in the source text: blanked out names are
# represented like "E----", which causes the language model to follow a hyphen
# more hyphens, which can sometimes spiral out of control into long sequences
# of nothing but hyphens. This crude regex is meant to curb that problem by
# reducing most sequences of hyphens down to only two consecutive hyphens.
```

```python
corpus_file = open('train.txt', 'w')
corpus_file.write(corpus)
```
